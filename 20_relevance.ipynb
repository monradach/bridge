## Function

```python
def typhoon_call(system_message, prompt):
    url = "https://api.together.xyz/v1/chat/completions"
    headers = {
        "Authorization": f"Bearer {api_key}",
        "Content-Type": "application/json"
    }
    data = {
        "model": "scb10x/scb10x-llama3-1-typhoon2-8b-instruct",
        "messages": [
            {"role": "system", "content": system_message},
            {"role": "user", "content": prompt}
        ],
        "temperature": 0.3,
        "top_p": 0.7,
        "repetition_penalty": 1.1,
    }

    response = requests.post(url, headers=headers, json=data)
    try:
        return response.json()["choices"][0]["message"]["content"]
    except Exception as e:
        print("Error in API response:", e)
        print("Full response:", response.text)
        return None
```

```python
df_sample = df.iloc[:20]
df_dropped = df_sample.drop(df_sample.columns[[0, 2]], axis=1)
```

```python
df_dropped
```

```python
results =[]
for index, row in df_dropped.iterrows():
    prompt_row = f"Is this message talking about Ramathibodi Yes or No: {row[0]}"
    result = typhoon_call(system_message, prompt_row)
    if result is not None:
        results.append({
            "response": result
        })
    else:
        print(f"Skipping prompt due to error: {prompt_row}")

print(results)
```

```python
df_result = pd.DataFrame(results)
df_result
```

```python
df_complete = pd.concat([df_dropped,df_result],axis=1)
df_complete.head(3)
#df_complete.to_csv("20sample.csv", index=False)
```

```python
df_complete["Yes/No"] = df_complete["response"].str[:3]
df_split = df_complete.drop(df_complete.columns[1],axis=1)
df_split
```

```python
for index, row in df_split.iterrows():
    if row[1] == 'No.':
        df_split.at[index, df_split.columns[1]] = 'No'
df_split
```

```python
df_final = pd.DataFrame(columns=df_split.columns)
df_final = df_split[df_split[df_split.columns[1]] == 'Yes'].copy()
df_final = df_final.reset_index(drop=True)
df_final = df_final.drop("Yes/No",axis=1)
df_final
```
