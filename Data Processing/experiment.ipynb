```python
def normalize_column_all(df, column_name):
    normalized_dfs = []
    
    for index, row in df.iterrows():
        # Extract the list of dictionaries for each row
        list_of_dicts = row[column_name]
        
        # Normalize the list of dictionaries into a DataFrame
        normalized_df = json_normalize(list_of_dicts)
        
        # Add the normalized DataFrame for the current row to the list
        normalized_dfs.append(normalized_df)
    
    # Concatenate all normalized DataFrames row-wise (with the same structure)
    return pd.concat(normalized_dfs, ignore_index=True)

# Normalize 'topPosts' and 'latestPosts' columns
top_posts_df = normalize_column_all(df, 'topPosts')
latest_posts_df = normalize_column_all(df, 'latestPosts')

# Combine the normalized 'topPosts' and 'latestPosts' DataFrames into the original DataFrame
result_df = pd.concat([df.drop(['topPosts', 'latestPosts'], axis=1), top_posts_df, latest_posts_df], axis=1)

# Display the result
print(result_df)
```
